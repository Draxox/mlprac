```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_iris

iris=load_iris()

iris

iris.data[:10]


iris.target

import seaborn as sns

df = sns.load_dataset('iris')

df

#independent and dependent features
x = df.iloc[:,:-1]
y = iris.target

x

#train test split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=52
                                                    )

from sklearn.tree import DecisionTreeClassifier

##post prunning

treemodel = DecisionTreeClassifier()

treemodel.fit(x_train, y_train)

from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)

y_pred = treemodel.predict(x_test)

y_pred

treemodel.score(x_test, y_test)

from sklearn.metrics import accuracy_score, classification_report

score=accuracy_score(y_test, y_pred)
print(score)

print(classification_report(y_pred, y_test))

#new code on new dataset






import pandas as pd

dff = pd.read_csv('salaries.csv')
dff.head()

X = dff.drop(['salary_more_then_100k'], axis = 1)
Y = dff['salary_more_then_100k']

from sklearn.preprocessing import LabelEncoder
le_comp = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()

X.company = le_comp.fit_transform(dff.company)
X.job = le_job.fit_transform(dff.job)
X.degree = le_degree.fit_transform(dff.degree)

X.head()

from sklearn import tree

ml = tree.DecisionTreeClassifier()

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.35, random_state = 42)

ml.fit(X_train, Y_train)

Y_pred = ml.predict(X_test)

from sklearn.metrics import accuracy_score
sc = accuracy_score(Y_test, Y_pred)
sc

ml.score(X_test, Y_test)

ml.predict([[2,2,1]])

Y_prob = ml.predict_proba(X_test)

from sklearn.metrics import confusion_matrix, classification_report
cr = classification_report(Y_test, Y_pred)
print(cr)

cn = confusion_matrix(Y_test, Y_pred)
cn

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (10,7))
sns.heatmap(cn, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()

from sklearn import tree
plt.figure(figsize = (10,7))
tree.plot_tree(ml, filled = True)
```
