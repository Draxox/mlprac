```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import accuracy_score, f1_score

# Load dataset
df = pd.read_csv('pima-indian-diabetes.csv')

# Features and target
X = df.drop('Outcome', axis=1)  # Assuming 'Outcome' is the label column
y = df['Outcome']

# Standardize features (important for GaussianNB)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)
acc_gnb = accuracy_score(y_test, y_pred_gnb)
f1_gnb = f1_score(y_test, y_pred_gnb)

print(f"GaussianNB Accuracy: {acc_gnb:.2f}")
print(f"GaussianNB F1-Score: {f1_gnb:.2f}")

# Multinomial Naive Bayes (typically used for discrete features)
mnb = MultinomialNB()
# MultinomialNB expects non-negative features, so use raw or scaled without negative transformation
X_train_mnb = X_train - X_train.min(axis=0)
X_test_mnb = X_test - X_train.min(axis=0)

mnb.fit(X_train_mnb, y_train)
y_pred_mnb = mnb.predict(X_test_mnb)
acc_mnb = accuracy_score(y_test, y_pred_mnb)
f1_mnb = f1_score(y_test, y_pred_mnb)

print(f"MultinomialNB Accuracy: {acc_mnb:.2f}")
print(f"MultinomialNB F1-Score: {f1_mnb:.2f}")
```
