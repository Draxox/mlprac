```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Cust_segmentation.csv')

# Encode the Address feature (categorical)
le = LabelEncoder()
df['Address_enc'] = le.fit_transform(df['Address'])

# Select features
X = df[['CustomerId', 'Address_enc']]

# Standardize features for better clustering results
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering (Euclidean distance)
k = 3  # number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)

# Add cluster labels to dataframe
df['Cluster'] = kmeans.labels_

# Plot clusters (2D)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.xlabel('CustomerId (scaled)')
plt.ylabel('Address (encoded & scaled)')
plt.title('K-means Clustering (Euclidean Distance)')
plt.show()
```
